```python
import numpy as np 
import matplotlib.pyplot as plt

ask =  int(input("Enter how much quatity do u want to calculate ? "))

ran= np.random.randint(1,100,ask)

mean= np.mean(ran)
std_dev=np.std(ran)
median = np.median(ran)
variance = np.var(ran)

print(mean)
print(median)
print(std_dev)
print(variance)
plt.figure(figsize=(10, 6))

# Histogram
plt.hist(ran, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='-.', label=f'Median: {median:.2f}')
plt.title("Random Numbers Distribution")
plt.xlabel("Number Range")
plt.ylabel("Frequency")
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
# print(ran)
```

```python
# np.random.seed(5)
# np.random.randint(1,10,20)
# np.random.seed(10)
# np.random.randint(1,10,20)

np.random.seed(10)
matrix_1=np.random.randint(1,10,(4,4))
matrix_2=np.random.randint(1,10,(4,4))

add=np.add(matrix_1,matrix_2)
sub=np.subtract(matrix_1,matrix_2)
mul=np.dot(matrix_1,matrix_2)

print(add)
print(sub)
print(mul)

plt.figure(figsize=(12, 8))

# Plot Matrix A
plt.subplot(2, 3, 1)
plt.imshow(matrix_1, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Matrix A")

# Plot Matrix B
plt.subplot(2, 3, 2)
plt.imshow(matrix_2, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Matrix B")

# Plot Matrix Sum
plt.subplot(2, 3, 3)
plt.imshow(add, cmap='coolwarm', aspect='auto')
plt.colorbar()
plt.title("Matrix A + Matrix B")

# Plot Matrix Difference
plt.subplot(2, 3, 4)
plt.imshow(sub, cmap='coolwarm', aspect='auto')
plt.colorbar()
plt.title("Matrix A - Matrix B")

# Plot Matrix Product
plt.subplot(2, 3, 5)
plt.imshow(mul, cmap='coolwarm', aspect='auto')
plt.colorbar()
plt.title("Matrix A x Matrix B")

plt.tight_layout()
plt.show()
```
